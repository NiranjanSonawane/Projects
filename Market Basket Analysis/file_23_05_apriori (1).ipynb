```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
final_data = df.groupby(['Transaction','Item'])['Item'].count().reset_index(name='count')
```

```python
final_data
```

```python
data_final = final_data.pivot_table(index='Transaction',columns='Item',values='count').fillna(0.0)
data_final.loc[2].value_counts()
```

```python
def change(s):
    if s<=0.0:
        return 0
    elif s>=1.0:
        return 1
basket_analaysis = data_final.applymap(change)
```

```python
basket_analaysis
```

```python
frequent_items  = apriori(basket_analaysis,min_support=0.05,low_memory=False,use_colnames=True)
frequent_items
```

```python
association_rules(frequent_items,metric='confidence',min_threshold=0.1,support_only=False)
```
